Matplotlib Basics

```python
import matplotlib.pyplot as plt
# Data
subjects = ["Math", "Physics", "Chemistry", "Biology",
"English"]
marks = [85, 90, 78, 88, 76]
# Plot
plt.plot(subjects, marks, marker='o', color='blue')
# Customization
plt.title("Student Marks in Different Subjects")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.grid(True)
# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt
# Data
days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
temperature = [30, 32, 31, 29, 33, 35, 34]
# Plot
plt.plot(days, temperature, marker='s', linestyle='--',
color='red')
# Customization
plt.title("Weekly Temperature Trend")
plt.xlabel("Days")
plt.ylabel("Temperature (C)")
plt.grid(True)
# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt
# Data
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
sales = [15000, 18000, 12000, 22000, 24000, 20000]
# Plot
plt.bar(months, sales, color='green')
# Customization
plt.title("Monthly Sales Report")
plt.xlabel("Months")
plt.ylabel("Sales in USD")
plt.grid(axis='y')
# Show plot
plt.show()

```

```python
import matplotlib.pyplot as plt
# Data
subjects = ["Math", "Physics", "Chemistry", "Biology",
"English"]
marks = [85, 90, 78, 88, 76]
# Plot
plt.plot(subjects, marks, marker='o', color='blue', linestyle='-')
# Labels and Title
plt.title("Student Marks")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.grid(True)
# Show
plt.show()
```

```python
import matplotlib.pyplot as plt
days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
temperature = [30, 32, 31, 29, 33, 35, 34]
plt.plot(days, temperature, marker='s', linestyle='--',
color='red')
plt.title("Weekly Temperature Trend")
plt.xlabel("Days")
plt.ylabel("Temperature (C)")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
stock_price = [150, 152, 148, 155, 160, 158, 162]
plt.plot(days, stock_price, marker='^', linestyle='-',
color='green', linewidth=2)
plt.title("Company Stock Price Movement")
plt.xlabel("Days")
plt.ylabel("Stock Price (in USD)")
plt.grid(True)
plt.show()

```

Seaborn for statistical visualization

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
data=np.random.rand(6,6)
corr=pd.DataFrame(data,columns=list("ABCDEF")).corr()
plt.figure(figsize=(8,6))
sns.heatmap(corr,annot=True,cmap="coolwarm",linewidth=0.5)
plt.title("Heatmap example-Correlation Matrix")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
df=sns.load_dataset("iris")
sns.pairplot(df,hue="species",diag_kind="kde",palette="Set2")
plt.suptitle("Pairplot example-iris dataset,y=1.02")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
tips=sns.load_dataset("tips")
plt.figure(figsize=(6,4))
sns.barplot(x="day",y="total_bill",data=tips,palette="muted")
plt.title("Barplot example-tips dataset")
plt.show()
plt.figure(figsize=(6,4))
sns.boxplot(x="day",y="total_bill",data=tips,palette="Set2")
plt.title("Boxplot example-tips dataset")
plt.show()
plt.figure(figsize=(6,4))
sns.violinplot(x="day",y="total_bill",data=tips,palette="Set3")
plt.title("violin example-tips dataset")
plt.show()
plt.figure(figsize=(6,4))
sns.countplot(x="day",data=tips,palette="coolwarm")
plt.title("countplot example-tips dataset")
plt.show()

```
